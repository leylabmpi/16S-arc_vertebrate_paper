## 16S-arch

```r
chunck = function(x, n=4){
    split(x, factor(1:length(x)%%n))
}
chunck(LETTERS, n=3)
```

```r
.classification = function(x, db, batch){
    res = classification(x, db=db, batch=batch)
    Sys.sleep(8)
    return(res)
}

classification_batch = function(x, db = 'ncbi', batch=8){
    n_c = as.integer(length(x) / batch)
    cls = chunck(x, n_c) %>%
        lapply(.classification, db = 'ncbi', batch=batch) %>%
        do.call(c, .)
    return(cls)
}
```

```r
# query NCBI
meta_cls = meta %>%
    mutate(scientific_name = gsub('_', ' ', scientific_name)) %>%
    .$scientific_name %>%
    classification_batch
    
meta_cls %>% length
```

# Overlap

```r
tax_overlap = function(cls, xlab, ylab, lev='species', ...){
    x = cls[[xlab]] %>%
        filter(rank == lev) %>%
        distinct(name)
    y = cls[[ylab]]%>%
        filter(rank == lev) %>%
        distinct(name)
    overlap(x$name, y$name, ...)
}
```
