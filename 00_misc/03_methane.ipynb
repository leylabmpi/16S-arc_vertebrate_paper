# Goal

* Overlap of published CH4 datasets & 16S-arch host species

# Var

```r
work_dir = '/ebio/abt3_projects/Georg_animal_feces/data/16S_arch/MiSeq-Runs-116-122-126-189-190/LLA/merged/'

# public data
## Hackstein & van Alen 1996
Hackstein_file = '/ebio/abt3_projects/Georg_animal_feces/data/animal/methane/Hackstein-vanAlen1996.xlsx'
## Clauss et al., 2020
Clauss_file = '/ebio/abt3_projects/Georg_animal_feces/data/animal/methane/Clauss-2020_methane_animals.xlsx'

# microbiome
## 16S-arch phyloseq object (one sample per species)
physeq_arch_SpecD_file = '/ebio/abt3_projects/Georg_animal_feces/data/16S_arch/MiSeq-Runs-116-122-126-189-190/LLA/merged/phyloseq/physeq_arch_SpecD.RDS'
## 16S-arch phyloseq object (one sample per individual)
physeq_arch_IndD_file = '/ebio/abt3_projects/Georg_animal_feces/data/16S_arch/MiSeq-Runs-116-122-126-189-190/LLA/merged/phyloseq/physeq_arch_IndD.RDS'

# alpha diversity
alpha_div_file = '/ebio/abt3_projects/Georg_animal_feces/data/16S_arch/MiSeq-Runs-116-122-126-189-190/LLA/merged/diversity-qiime2/alpha_div_meth.RDS'

# host tree
host_tree_SpecD_file = '/ebio/abt3_projects/Georg_animal_feces/data/animal/phylogeny/TT/unified_metadata_complete_20190416_rn.nwk'
host_tree_IndD_file = '/ebio/abt3_projects/Georg_animal_feces/data/animal/phylogeny/TT/unified_metadata_complete_200429_16S-arch_IndD.nwk'

# params
threads = 8
my_seed = 8371
```

# Load

## 16S-arch

```r
# archaea
physeq_arch_IndD = readRDS(physeq_arch_IndD_file)
physeq_arch_IndD
```

```r
# metadata
meta = physeq_arch_IndD %>%
    phyloseq2df(sample_data) %>%
    dplyr::select(SampleID, class, order, family, genus, scientific_name,
                  diet, habitat) %>%
    mutate(scientific_name = gsub('_', ' ', scientific_name))
meta 
```

```r
# alpha diversity
alpha_div = readRDS(alpha_div_file) %>%
    do.call(cbind, .) %>%
    mutate(Sample = rownames(.))
alpha_div
```

## Public metadata

```r
Hackstein = readxl::read_excel(Hackstein_file) %>%
    mutate(genus = gsub(' .+', '', Scientific_name)) 
Hackstein
```

```r
Clauss = readxl::read_excel(Clauss_file, sheet='methane') %>%
    group_by(Species) %>%
    summarize(CH4_mean = mean(`CH4_g/d`, na.rm=TRUE),
              .groups = 'drop') %>%
    mutate(genus = gsub(' .+', '', Species))
Clauss
```

# Combining

```r
CH4 = meta %>%
    left_join(Hackstein, c('scientific_name'='Scientific_name')) %>%
    left_join(Clauss, c('scientific_name' = 'Species')) %>%
    dplyr::select(SampleID, class, order, family, genus.x,
                  scientific_name, diet, CH4_mean.x, CH4_max, H2,
                  CH4_mean.y) %>%
    mutate(CH4_mean = ifelse(!is.na(CH4_mean.y), CH4_mean.y, CH4_mean.x),
           study = ifelse(!is.na(CH4_mean.y), 'Clauss_2020', 'Hackstein_1996')) %>%
    group_by(SampleID, class, order, family, genus.x,
                  scientific_name, diet, study) %>%
    summarize(CH4_mean = mean(CH4_mean, na.rm=TRUE),
              .groups='drop') %>%
    filter(!is.na(CH4_mean))
CH4 
```

```r
CH4 %>%
    distinct(study, scientific_name) %>%
    .$study %>% table
```

# Mapping CH4 onto the host tree

## metadata

```r
x = CH4 %>%
    filter(!is.na(CH4_mean)) %>%
    group_by(scientific_name) %>%
    summarize(CH4_mean = mean(CH4_mean, na.rm=TRUE),
              .groups='drop') %>%
    as.data.frame
rownames(x) = gsub(' ', '_', x$scientific_name)
x$scientific_name = NULL
x
```

```r
out_file = file.path(work_dir, 'SpecD_CH4_iTOL-bar.txt')
itol_simplebar(x, 'CH4', out_file, WIDTH=100)
```

## Tree

```r
tree = read.tree(host_tree_SpecD_file)
tree
```

```r
to_rm = setdiff(tree$tip.label, rownames(x))
tree_f = ape::drop.tip(tree, to_rm)
tree_f
```

```r
out_file = file.path(work_dir, 'SpecD_CH4.nwk')
write.tree(tree_f, out_file)
cat('File written:', out_file, '\n')
```

### Methanogen diversity

```r
x = alpha_div %>%
    semi_join(CH4, c('Sample'='SampleID')) %>%
    dplyr::select(-pielou_e, -observed_otus) %>%
    inner_join(meta %>% dplyr::select(SampleID, scientific_name),
               c('Sample'='SampleID')) %>%
    mutate(scientific_name = gsub(' ', '_', scientific_name)) %>%
    group_by(scientific_name) %>%
    summarize(shannon = mean(shannon, na.rm=TRUE),
              faith_pd = mean(faith_pd, na.rm=TRUE),
              .groups='drop') %>%
    as.data.frame
rownames(x) = x$scientific_name
x$scientific_name = NULL

out_file = file.path(work_dir, 'SpecD_methAlpha_itol-bar.txt') 
itol_multibar(x, 'meta_alphaDiv', out_file)
```

# Association with diversity

```r
overlap(CH4$SampleID, alpha_div$Sample)
```

```r
# joining
CH4_j = CH4 %>%
    inner_join(alpha_div, c('SampleID'='Sample')) 
CH4_j
```

```r
# summarizing by species
CH4_js = CH4_j %>%
    filter(!is.na(CH4_mean)) %>%
    group_by(scientific_name, class, order, family, genus.x, diet) %>%
    summarize(`Shannon Index` = mean(shannon),
              `Faith's PD` = mean(faith_pd),
              mean_CH4 = mean(CH4_mean),
              .groups='drop') %>%
    gather(metric, value, -mean_CH4, -scientific_name, -class, 
           -order, -family, -genus.x, -diet)
CH4_js
```

```r
# plotting
p = CH4_js %>%
    ggplot(aes(value, mean_CH4)) +
    geom_point(aes(shape=class, color=diet)) +
    geom_smooth() +
    labs(x='Alpha div.', y='CH4 (g/d)') +
    facet_wrap(~ metric, scales='free_x') +
    theme_bw()

p.dims(7,3)
plot(p)
```

```r
# plotting
p = CH4_js %>%
    filter(class != 'Amphibia') %>%
    ggplot(aes(value, mean_CH4, color=class)) +
    geom_point(aes(shape=diet)) +
    geom_smooth(method='lm', se=FALSE) +
    labs(x='Alpha div.', y='CH4 (g/d)') +
    facet_wrap( ~ metric, scales='free_x') +
    theme_bw()

p.dims(8,3)
plot(p)
```

```r
# plotting
p = CH4_js %>%
    mutate(is_herbivore = diet == 'herbivore') %>%
    ggplot(aes(value, mean_CH4)) +
    geom_point(aes(color=diet, shape=class)) +
    geom_smooth(method='lm') +
    labs(x='Alpha div.', y='CH4 (g/d)') +
    facet_grid(is_herbivore ~ metric, scales='free_x') +
    theme_bw()

p.dims(7,3.5)
plot(p)
```

```r
# plotting
p = CH4_js %>%
    filter(class != 'Amphibia') %>%
    mutate(is_herbivore = diet == 'herbivore') %>%
    ggplot(aes(value, mean_CH4, color=class)) +
    geom_point(aes(shape=diet)) +
    geom_smooth(method='lm') +
    labs(x='Alpha div.', y='CH4 (g/d)') +
    facet_grid(is_herbivore ~ metric, scales='free_x') +
    theme_bw()

p.dims(7,3.5)
plot(p)
```

## Stats

```r
# summarizing by species
CH4_js = CH4_j %>%
    filter(!is.na(CH4_mean)) %>%
    group_by(scientific_name, class, order, family, genus.x, diet) %>%
    summarize(mean_shannon = mean(shannon),
              mean_faith_pd = mean(faith_pd),
              mean_CH4 = mean(CH4_mean),
              .groups='drop') %>%
    gather(metric, value, -mean_CH4, -scientific_name, -class, 
           -order, -family, -genus.x, -diet)
CH4_js
```
