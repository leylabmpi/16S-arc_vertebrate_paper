## Microbiome

### Number of ASVs (OTUs

```r
# taxonomy
tax = physeq_arch_IndD %>%
    phyloseq2df(tax_table) %>%
    as.data.table

tax
```

```r
tax_s = tax %>%
    group_by(Phylum) %>%
    summarize(n_OTUs = n()) %>%
    ungroup()

df.dims(30)
tax_s %>% arrange(-n_OTUs)
df.dims()
```

```r
tax_s = tax %>%
    group_by(Phylum, Class) %>%
    summarize(n_OTUs = n()) %>%
    ungroup()

df.dims(30)
tax_s %>% arrange(-n_OTUs)
df.dims()
```

```r
# plotting
p = tax_s %>%
    mutate(N_OTUs = ifelse(n_OTUs == 1, 1.2, n_OTUs)) %>%
    mutate(Class = Class %>% reorder(-n_OTUs)) %>%
    ggplot(aes(Class, N_OTUs, fill=Phylum)) +
    geom_bar(stat='identity') +
    geom_text(aes(label=n_OTUs, y=N_OTUs * 1.3), size=3.2) +
    scale_y_log10() +
    labs(y='No. of ASVs') +
    theme_bw() +
    theme(
        axis.text.x = element_text(angle=45, hjust=1)
    )

p.dims(5,3)
plot(p)
```

### Taxon abundances

```r
otu = physeq_arch_IndD %>%
    phyloseq2df(otu_table, long=TRUE) %>%
    as.data.table

tax = physeq_arch_IndD %>%
    phyloseq2df(tax_table) %>%
    as.data.table

otu = otu %>%
    dt_inner_join(tax, c('OTU')) %>%
    dt_mutate(total_count = sum(Count), by=c(Sample)) %>%
    dt_mutate(perc_abund = Count / total_count * 100) %>%
    dt_select(-total_count)
rm(tax)

otu
```

```r
# adding metadata
otu = otu %>%
    dt_inner_join(meta, c('Sample' = 'SampleID'))
otu
```

#### Phylum

```r
# summarizing by taxonomy
otu_s = otu %>%
    dt_mutate(Taxonomy = paste(Phylum, sep=';')) %>%
    dt_summarize(perc_abund = sum(perc_abund),
                 by = c(Taxonomy, Sample, class, order, family, genus, scientific_name, diet))
otu_s
```

```r
# rare clades
otu_s %>%
    dt_filter(Taxonomy == 'Asgardaeota' | Taxonomy == 'Diapherotrites') %>%
    dt_filter(perc_abund > 0)
```

```r
# plotting
p = otu_s %>%
    dt_mutate(perc_abund = ifelse(perc_abund <= 0, 1e-3, perc_abund)) %>%
    dt_mutate(Taxonomy_abund = sum(perc_abund), by=c(Taxonomy)) %>%
    dt_mutate(Taxonomy = Taxonomy %>% reorder(-Taxonomy_abund)) %>%
    ggplot(aes(class, perc_abund)) +
    geom_boxplot(outlier.size=0.7) +
    scale_y_log10(breaks=c(0.001, 0.01, 0.1, 1, 10, 100),
                  labels = function(x) format(x, scientific = FALSE)) +
    labs(x='Host class', y='% abundance') +
    facet_wrap(~ Taxonomy) +
    theme_bw() +
    theme(
        axis.text.x = element_text(angle=45, hjust=1)
    )

p.dims(5.5,3.8) 
plot(p)
```

### Beta dispersion

```r
dist_mtx_order = function(d, x){
    # Ordering distance matrixes
    # d = distance matrix (dist class)
    # x = vector to order dist. obj. by
    m = d %>% as.matrix
    d = as.dist(m[x,x])
    return(d)
}
```

```r
beta_disp = function(d, fac, samples){
    # checking beta dispersion
    d = dist_mtx_order(d, samples)
    ret = vegan::betadisper(d,fac)
    ret = data.frame(group = ret$group,         # note: group is in wrong order; but can just match host_subj_id based on sampleID
                     distances = ret$distances)
    return(ret)
}
```

```r
format_data = function(x){
    # format output
    x = do.call(rbind, x) 
    x$metric = gsub('\\..+', '', rownames(x))
    x$sample_id = gsub('^.*?\\.', '', rownames(x))
    return(x)
}
```

```r
.beta_disp_all = function(group, d, metadata, threads=2){
    if(! group %in% colnames(metadata)){
        stop(glue::glue('Cannot find {G} in metadata', G=group))
    }
    doParallel::registerDoParallel(cores=threads)
    ret = plyr::llply(d, beta_disp, fac=metadata[,group], 
                      samples=rownames(metadata), .parallel=TRUE)
    ret = format_data(ret)
    ret$grouping = group %>% as.character
    return(ret)
}

beta_disp_all = function(list_of_dists, metadata, groups, threads=2){
    groups = as.list(groups)
    ret = lapply(groups, .beta_disp_all, d=list_of_dists, metadata=metadata, threads=threads)
    ret = do.call(rbind, ret)
    return(ret)
}
```
