## Summary

````r
# number of contigs
F = '/ebio/abt3_projects/Georg_animal_feces/data/metagenome/HiSeqRuns-n11/LLMGA/v0.12/contigs_stats.tsv'
quast_info = Fread(F, fill=TRUE) %>%
    rename.('Sample' = file,
            'n_contigs' = num_seqs) %>%
    select.(Sample, n_contigs) %>%
    mutate.(Sample = gsub('^F', 'XF', Sample))
quast_info
````

````r
# loading hits
F = file.path(work_dir, 'minimap2', 'LCA_sr-relaxed_cov0.7.tsv')
mm = Fread(F) %>%
    separate.(lca_lineage, c('root', taxonomy_levels()), sep=';') %>%
    select.(-root)
mm
````

````r
# hits per contig per sample
mm %>%
    summarize.(n = n.(),
               .by=c(query_id, Name)) %>%
    .$n %>% summary_x('hits per contig per sample')
````

````r
# checking sample overlap
overlap(mm$Name, meta$SampleID)
````

````r
# coverage (aln_len / query_len)
summary_x(mm$query_coverage)
````

````r
# filtering by coverage
mm = mm %>%
    filter.(query_coverage >= 0.9)
summary_x(mm$query_coverage)
````

````r
# merging with n-contig info
overlap(mm$Name, quast_info$Sample)
mm = mm %>%
    inner_join.(quast_info, c('Name'='Sample'))
mm
````

### Full taxonomy

````r
mm %>% 
    summarize.(n = n.(),
               .by=c(Domain, Phylum, Class))
````

````r
df.dims(10)
mm %>% 
    summarize.(n = n.(),
               .by=c(Domain, Phylum, Class, Order, Family)) 
df.dims()
````

#### By sample

````r
mm_s = mm %>% 
    summarize.(n = n.(),
               cov = mean(query_coverage),
               .by=c(Domain, Phylum, Class, Order, Family, Name, n_contigs)) %>%
    inner_join.(meta, c('Name' = 'SampleID')) %>%
    mutate.(perc = n / n_contigs * 100)
mm_s
````

````r
# plotting hits
p = mm_s %>% 
    mutate.(m_tax = paste(Order, Family, sep=';')) %>%
    ggplot(aes(class, n, color=m_tax)) +
    geom_boxplot() +
    scale_y_log10() +
    labs(y='No. of contigs') +
    theme_bw()
p.dims(8.5,3)
plot(p)
````

````r
# plotting hits
p = mm_s %>% 
    mutate.(m_tax = paste(Order, Family, sep=';')) %>%
    ggplot(aes(class, perc, color=m_tax)) +
    geom_boxplot() +
    labs(y='% of contigs') +
    theme_bw()
p.dims(8.5,3)
plot(p)
````

````r
# plotting hits
p = mm_s %>% 
    mutate.(m_tax = paste(Order, Family, sep=';')) %>%
    ggplot(aes(class, perc, color=m_tax)) +
    geom_boxplot() +
    scale_y_log10() +
    labs(y='% of contigs') +
    theme_bw()
p.dims(8.5,3)
plot(p)
````

### Just M-thermo (species)

````r
mthermo_fam = c('f__methanothermobacteraceae', 'f__methanothermobacteraceae_a')
````

````r
mm_s = mm %>% 
    filter.(Family %in% mthermo_fam) %>%
    summarize.(n = n.(),
               cov = mean(query_coverage),
               .by=c(Genus, Species, Name, n_contigs)) %>%
    inner_join.(meta, c('Name' = 'SampleID')) %>%
    mutate.(perc = n / n_contigs * 100)
mm_s$n %>% summary_x('No. of contigs')
mm_s$perc %>% summary_x('% of contigs')
mm_s
````

#### No. of hits

````r
# plotting hits by class
p = mm_s %>% 
    mutate.(m_tax = paste(Genus, Species, sep=';')) %>%
    ggplot(aes(class, perc, color=m_tax)) +
    geom_boxplot() +
    scale_y_log10() +
    labs(y='% of contigs') +
    theme_bw()
p.dims(8.5,3)
plot(p)
````

````r
# plotting hits by Aves order
p = mm_s %>% 
    filter.(class == 'Aves') %>%
    mutate.(m_tax = paste(Species, sep=';')) %>%
    mutate.(m_tax = gsub('^s__', '', m_tax)) %>%
    mutate.(m_tax = gsub('^m', 'M', m_tax)) %>%
    ggplot(aes(order, perc, color=m_tax)) +
    geom_boxplot() +
    scale_y_log10() +
    labs(y='% of contigs') +
    scale_color_discrete('Genus-Species') +
    theme_bw() +
    theme(
        axis.text.x = element_text(angle=45, hjust=1)
    )
p.dims(8,3)
plot(p)
````

#### Coverage

````r
# plotting hits by class
p = mm_s %>% 
    mutate.(m_tax = paste(Genus, Species, sep=';')) %>%
    ggplot(aes(class, cov, color=m_tax)) +
    geom_boxplot() +
    theme_bw() +
    theme(
        axis.text.x = element_text(angle=45, hjust=1)
    )
p.dims(8.5,3)
plot(p)
````

````r
# plotting hits by Aves order
p = mm_s %>% 
    filter.(class == 'Aves') %>%
    mutate.(m_tax = paste(Genus, Species, sep=';')) %>%
    ggplot(aes(order, cov, color=m_tax)) +
    geom_boxplot() +
    labs(y='Coverage') +
    theme_bw() +
    theme(
        axis.text.x = element_text(angle=45, hjust=1)
    )
p.dims(9,3)
plot(p)
````

### Just M-thermo (genus)

````r
# summarizing
mm_s = mm %>%
    filter.(Family %in% mthermo_fam) %>%
    mutate.(Genus = ifelse(Genus == 'g__methanothermobacter_a', 
                           'g__methanothermobacter', Genus)) %>%
    summarize.(n = n.(),
               cov = mean(query_coverage),
               .by=c(Genus, Name, n_contigs)) %>%
    mutate.(perc_contigs = n / n_contigs * 100) %>%
    inner_join.(meta, c('Name' = 'SampleID'))
mm_s$n %>% summary_x('No. of contigs')
mm_s$perc %>% summary_x('% of contigs')
mm_s
````

````r
# plotting hits by Aves order
p = mm_s %>% 
    filter.(class == 'Aves') %>%
    ggplot(aes(order, n)) +
    geom_boxplot() +
    scale_y_log10() +
    labs(y='No. of contigs') +
    theme_bw() +
    theme(
        axis.text.x = element_text(angle=45, hjust=1)
    )
p.dims(5,3)
plot(p)
````

````r
# plotting hits by Aves sample
mm_s %>% filter.(class == 'Aves') %>% summary_x('No. of contigs', n)

p = mm_s %>% 
    filter.(class == 'Aves') %>%
    mutate.(n = ifelse(n == 1, 1.1, n)) %>%
    mutate.(Name = reorder(Name, n)) %>%
    ggplot(aes(Name, n, fill=order)) +
    geom_bar(stat='identity') +
    scale_y_log10() +
    labs(x='Sample', y='No. of contigs') +
    theme_bw() +
    coord_flip()
p.dims(5.5,8)
plot(p)
````

````r
# plotting hits by Aves sample
mm_s %>% filter.(class == 'Aves') %>% summary_x('% of contigs', perc_contigs)

p = mm_s %>% 
    filter.(class == 'Aves') %>%
    mutate.(Name = reorder(Name, perc_contigs)) %>%
    ggplot(aes(Name, perc_contigs, fill=order)) +
    geom_bar(stat='identity') +
    geom_text(aes(label=n), size=3.5, hjust=1) +
    scale_fill_discrete('Order') +
    labs(x='Sample', y='% of contigs') +
    theme_bw() +
    coord_flip()
p.dims(5.5,8.5)
plot(p)
````

````r
# plotting hits by Aves sample
mm_f = mm %>%
    filter.(Family %in% mthermo_fam) %>%
    inner_join.(meta, c('Name'='SampleID')) %>%
    filter.(class == 'Aves') %>%
    mutate.(n = n.(),
            cov = mean(query_coverage),
            .by=c(Genus, Name, n_contigs)) %>%
    mutate.(perc_contigs = n / n_contigs * 100) 

summary_x(mm_f$cov, 'coverage')

p = mm_f %>%
    mutate.(Name = reorder(Name, perc_contigs)) %>%
    ggplot(aes(Name, query_coverage, color=order)) +
    geom_boxplot() +
    scale_y_log10() +
    scale_fill_discrete('Order') +
    labs(x='Sample', y='Query coverage') +
    theme_bw() +
    coord_flip()
p.dims(5.5,8)
plot(p)
````
