# Estimating coverage

```r
# rarefaction-based coverage estimation
doParallel::registerDoParallel(threads)
rarefaction_curve_data = calculate_rarefaction_curves(physeq_arch, 
                                                      c('Observed', 'Shannon'), 
                                                      rep(c(10, 50, 100, 250, 500, 1:10 * 1000), 
                                                      each = 100), 
                                                      parallel = TRUE) 

rarefaction_curve_data
```

```r
# summarizing
rarefaction_curve_data = rarefaction_curve_data %>% 
    group_by(Depth, Sample, Measure) %>%
    summarize(mean_div = mean(Alpha_diversity, na.rm=TRUE),
              std_div = sd(Alpha_diversity, na.rm=TRUE)) 

rarefaction_curve_data
```

```r
p = rarefaction_curve_data %>%
    ggplot(aes(Depth, mean_div,
               ymin = mean_div - std_div,
               ymax = mean_div + std_div,
               group = Sample)) + 
    geom_line(alpha=0.1) +
    geom_linerange(alpha=0.2) +
    geom_point(alpha=0.2) +
    scale_x_log10(breaks=c(10, 100, 250, 500, 1000, 2000, 5000, 10000)) +
    labs(x='No. of sequences', y='Value') +
    facet_grid(Measure ~ ., scales = 'free_y') +
    theme_bw()
      
p.dims(6,4)
Plot(p)
```

```r
p = rarefaction_curve_data %>%
    inner_join(meta_f %>% dplyr::select(SampleID, class), 
               c('Sample'='SampleID')) %>%
    ggplot(aes(Depth, mean_div,
               ymin = mean_div - std_div,
               ymax = mean_div + std_div,
               group = Sample)) + 
    geom_line(alpha=0.1) +
    geom_linerange(alpha=0.2) +
    geom_point(alpha=0.2) +
    scale_x_log10(breaks=c(10, 100, 250, 500, 1000, 2000, 5000, 10000)) +
    labs(x='No. of sequences', y='Value') +
    facet_grid(Measure ~ class, scales = 'free_y') +
    theme_bw() +
    theme(
        axis.text.x = element_text(angle=90, hjust=1, vjust=0.5)
    )
      
p.dims(9,4)
Plot(p)
```

```r
p = rarefaction_curve_data %>%
    filter(Measure == 'Shannon') %>%
    ggplot(aes(Depth, mean_div)) + 
    geom_point(alpha=0.2) +
    geom_smooth() +
    scale_x_log10(breaks=c(10, 100, 500, 1000, 2000, 5000, 10000)) +
    labs(x='No. of sequences', y='Value') +
    facet_grid(Measure ~ ., scales = 'free_y') +
    theme_bw()
      
p.dims(6,4)
plot(p)
```

# Summarizing coverage

```r
# WARNING: somewhat slow
doParallel::registerDoParallel(threads)
rarefaction_curve_data = calculate_rarefaction_curves(physeq_arch_f, 
                                                      c('Observed', 'Shannon'), 
                                                      rep(c(10, 25, 50, 100, 250, 500, 1:10 * 1000),
                                                          each=100),
                                                      parallel=TRUE) 

rarefaction_curve_data
```

```r
# summarizing
rarefaction_curve_data = rarefaction_curve_data %>% 
    group_by(Depth, Sample, Measure) %>%
    summarize(mean_div = mean(Alpha_diversity, na.rm=TRUE),
              std_div = sd(Alpha_diversity, na.rm=TRUE)) 

rarefaction_curve_data
```

```r
rarefaction_curve_data = rarefaction_curve_data %>%
    inner_join(meta %>% as.matrix %>% as.data.frame %>% 
               dplyr::select(SampleID, class, order, family, genus, diet),
                c('Sample'='SampleID')) 

rarefaction_curve_data
```

```r
p = rarefaction_curve_data %>%
    ggplot(aes(Depth, mean_div,
                ymin = mean_div - std_div,
                ymax = mean_div + std_div,
                group = Sample)) + 
    geom_line(alpha=0.1) +
    geom_linerange(alpha=0.2) +
    geom_point(alpha=0.2, size=0.8) +
    scale_x_log10(breaks=c(10, 25, 50, 100, 250, 500, 1000, 2000, 5000, 10000))+
    labs(x='Rarefaction depth (No. of sequences)', y='') +
    facet_grid(Measure ~ ., scales = 'free_y') +
    theme_bw() +
    theme(
        axis.text.x = element_text(angle=45, hjust=1)
    )
      
p.dims(5.5,4)
Plot(p)
```
